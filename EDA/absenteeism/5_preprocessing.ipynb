```python
#lecture30: get dummies
reason_columns = pd.get_dummies(df["Reason for Absence"])
#reason_columns
```

```python
reason_columns["check"] = reason_columns.sum(axis=1)
#reason_columns
```

```python
print(reason_columns["check"].sum(axis=0))
print(reason_columns["check"].unique())
```

```python
reason_columns = reason_columns.drop(["check"],axis=1)
#reason_columns
```

```python
reason_columns = pd.get_dummies(df["Reason for Absence"],drop_first=True)
#reason_columns
```

```python
#lecture35: grouping
print(df.columns.values) #names of columns
print(reason_columns.columns.values)
```

```python
df = df.drop("Reason for Absence", axis=1)
```

```python
#grouping the reason for absence
#group1: 1-14
#group2: 15-17
#group3: 18-21
#group4: 22-28
reason_columns.loc[:,'1':'14'] # = reason_columns.loc[:,1:14]
```

```python
#reason_columns.loc[:,1:14].max(axis=1)
reason_type_1 = reason_columns.loc[:,1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:,15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:,18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:,22:28].max(axis=1)
```

```python
#lecture36: concatenation
df = pd.concat([df,reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis=1)
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1','Reason_2','Reason_3','Reason_4']
```

```python
df.columns = column_names
df.head()
```

```python
#lecture39: reorder
column_names_reordered = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours','Reason_1','Reason_2','Reason_3','Reason_4']
```

```python
#lecture42: checkpoints
#creating a copy of the current state of the df Dataframe
df_checkpoint = df.copy()
df_checkpoint
```

```python
#lecture45: Date Format
#print(type(df_checkpoint['Date'][6]))
df_checkpoint['Date'] = pd.to_datetime(df_checkpoint['Date'], format = '%d/%m/%Y')
df_checkpoint['Date']
```

```python
df_checkpoint.info()
```

```python
#lecture46: month
print(df_checkpoint['Date'][0])
df_checkpoint['Date'][0].month
```

```python
list_months = []
for i in range(df_checkpoint.shape[0]): list_months.append(df_checkpoint['Date'][i].month)

df_checkpoint['Month Value'] = list_months
```

```python
df_checkpoint
```

```python
#lecture47: day of the week
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_checkpoint['Day of the Week'] = df_checkpoint['Date'].apply(date_to_weekday)

df_checkpoint.head()
```

```python
#lecture50: education
df_checkpoint2 = df_checkpoint.copy()
```

```python
df_checkpoint2['Education'].value_counts()
```

```python
df_checkpoint2['Education'] = df_checkpoint2['Education'].map({1:0,2:1,3:1,4:1})
print(df_checkpoint2['Education'].unique())
print(df_checkpoint2['Education'].value_counts())
```
